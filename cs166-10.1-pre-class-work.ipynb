```python
import numpy.random as rd
from matplotlib import pyplot as plt
import numpy as np

p_h = 0.6
p_t = 0.4

def random_coin_flip():
    location = 0
    for _ in range(30):
        flip = rd.random()
        if flip <= p_h: #gets heads
            location += 1 #go right
        else:
            location -= 1 #go left
    return (location)

hist_data = []

for _ in range(200):
    hist_data.append(random_coin_flip())
    
print ("Sample mean:" , np.mean(hist_data))
print ("Sample variance:", np.var(hist_data))

plt.hist(hist_data)
plt.show()
    
```

```python
walk = []
def random_coin_flip_walk():
    location = 0
    for _ in range(30):
        flip = rd.random()
        if flip <= p_h: #gets heads
            location += 1 #go right
            walk.append(1)
        else:
            location -= 1 #go left
            walk.append(-1)
    return (walk)

walk = random_coin_flip_walk()

y = np.zeros(31)
for i in range(31):
    y[i] = y[i-1] + walk[i-1]

plt.plot(np.linspace(1,31,31),y)
plt.xlabel("Time")
plt.ylabel("Movement (positive is to the right)")
plt.show()
```

```python
import random 

i_values = [10000, 20000, 50000, 100000, 500000]
for i_value in i_values:
    R = [] 
    i = 0
    R.append(100) # gambler’s initial fortune

    while R[i] > 0 and R[i] < 2100:
        i = i + 1
        if i >= i_value:
            break
        W = random.randint(0,1) #random value of 0 or 1
        W = 2 * W - 1 #random value of +/-1
        R.append(R[i - 1] + W) #gamblers new fortune

    plt.plot(range(len(R)), R) #plot R against its index
    plt.show()
    print ("Mean:", np.mean(R))
    print ("Variance:", np.var(R))
```
